```python
#makes ordered set starting at 1 that doubles until below largest num.
def make_set():
    doubled_numbers = OrderedSet()
    big_n = 1
    while (big_n <= largest):
        doubled_numbers.add(big_n)
        if(big_n*2 <= largest):
            tracker.append([big_n,big_n])
        big_n += big_n
    #print (doubled_numbers)
    return doubled_numbers
available_nums = make_set()
```

```python
def find_two_numbers(numbers_set, target_sum):
    """
    Find two numbers in a set that add up to a target sum.
    """
    # Create an empty set to store numbers we have seen so far
    seen_numbers = set()

    # Iterate through the numbers in the set
    for num in numbers_set:
        # Check if the difference between the target sum and the current number is in the set
        if target_sum - num in seen_numbers:
            # If so, return the two numbers that add up to the target sum
            return num, target_sum - num
        # Otherwise, add the current number to the set of seen numbers
        seen_numbers.add(num)

    # If no two numbers add up to the target sum, return None
    return None

```

```python
# given a desired_num and an OrderedSet it will find the value in the OrderedSet that is closest to the 
# desired_num that is less then desired_num
def find_close_num(desired_num, OrderedSet):
    if(desired_num in OrderedSet):
        return desired_num
    two_nums = find_two_numbers(OrderedSet, desired_num)
    if (two_nums is not None):
        return two_nums[0]
    if(desired_num % 2 == 0 and desired_num/2 in OrderedSet):
        return int(desired_num/2)
    
    best_diff = sys.maxsize
    closest_num = 1
    for number in OrderedSet:
        if(np.abs(desired_num-number)<best_diff and number <= desired_num):
            best_diff = np.abs(desired_num-number)
            closest_num = number
    return (closest_num)
```

```python
def optimize_path(small_tracker, big_num):
    size = len(small_tracker)
    if (size == 0):
        return

    if (big_num == 0):
        big_num = small_tracker[0][0]
    small_tracker2 = []
    sum = 0
  
    if (size <= 1):
        tracker.append(small_tracker[0])
        available_nums.add(small_tracker[0][0]+small_tracker[0][1])
        return
    for i in range(0,size-1,2):
        sum = small_tracker[i][1]+small_tracker[i+1][1]
        small_tracker2.append([big_num,sum])
        tracker.append([small_tracker[i][1],small_tracker[i+1][1]])
        available_nums.add(sum)
    if(size % 2 != 0):
        small_tracker2.append([sum,small_tracker[size-1][1]])
    optimize_path(small_tracker2, big_num)

    
```

```python
# get path for one numbers
def find_path2(number, step, small_tracker):
    close_num = (find_close_num(number, available_nums))
    close_num_2 = 0
    if (number != close_num):
        dif = number - close_num
        close_num_2 = (find_close_num(dif, available_nums))
    else:
        #print("number", number, "in list already")
        return
    #print ("step", step ,"to get number",number,"is:",close_num,close_num_2)

    small_tracker.append([close_num,close_num_2])
    
    sum = close_num+close_num_2
    available_nums.add(sum)
    #print ("sum", sum)
    step +=1
    if(sum != number):
        find_path2(number, step, small_tracker)
    else:
        #print("HIT")
        for i in range(0,step-1):
           #print("gone", available_nums.pop())
           available_nums.pop()
        optimize_path(small_tracker, 0)
```
